# What is this?
This is simple python script to crop the images taken by the camera. This is completely optional, but I found in my testing DRAMATICALLY increase the accuracy of the model and detection since it crops out the highly compressed/pixelated segment of the webcam frames.

```python
import cv2
# uncomment this line and the other plotting lines to see what the crop is cropping
# import matplotlib
# import matplotlib.pyplot as plt
```

```python
# this should be the directory with the images to crop
TO_CROP = "frames"
```

```python
def crop(im_path):
    crop_height = 400
    crop_width = 400
    image = cv2.imread(im_path)

    (height, width, _) = image.shape 
    # Draw a rectangle around the region you want to crop
    # cv2.rectangle(image, (0, height-crop_height), (crop_width, height), (255, 0, 0), 2)
    cropped_image = image[height-crop_height:, 0:crop_width]

    # plt.imshow(cropped_image)
    # plt.show()

    cv2.imwrite(im_path, cropped_image)
```

```python
from os import walk, path

# get all the images in the given directory
filenames = next(walk(TO_CROP), (None, None, []))[2]
filenames
```

```python
for filename in filenames:
    crop(path.join(TO_CROP, filename))
```
